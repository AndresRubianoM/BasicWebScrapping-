```python
def get_prices(flight):
    '''A function that returns a dictionary list with the prices'''
    prices = flight.find_elements_by_xpath('//table[@class = "fare-options-table"]//tfoot//td[contains(@class, "fare-")]')
    
    costs = []
    for price in prices:
        name = price.find_element_by_xpath('.//label').get_attribute('for')
        coin = price.find_element_by_xpath('.//span[@class = "currency-symbol"]').text
        val = price.find_element_by_xpath('.//span[@class = "value"]').text
        dict_info = {'Name': name, 'Coin':coin, 'Value': val}
        costs.append(dict_info)
        
    return costs
```

```python
def get_stops(flight):
    '''A function that returns a all the stops of the flight'''
    stops = driver.find_elements_by_xpath('//div[@class = "sc-jhAzac egnmB"]/div[@class ="sc-caSCKo emvtSU"]')
    stops_durations = flight.find_elements_by_xpath('//div[@class = "sc-caSCKo bSfcuj"]')
    
    info_stops = []
    stop_number = 0
    for stop in stops:
        
        #get cities
        cities = stop.find_elements_by_xpath('.//div[@class = "sc-gPEVay bRWLvO"]/abbr')
        city_out = cities[0].text
        city_in = cities[1].text
        #get hour out
        hours = stop.find_elements_by_xpath('.//div[@class = "sc-gPEVay bRWLvO"]/time')
        hour_out = hours[0].get_attribute('datetime')
        hour_in = hours[1].get_attribute('datetime')
        #get duration 
        duration = stop.find_element_by_xpath('.//span[@class = "sc-cvbbAY kxLrTU"]/time').get_attribute('datetime')
        #get airplane
        airplane = stop.find_element_by_xpath('.//div[@class = "airline-flight-details"]/span[@class = "sc-kEYyzF ktKVnp"]').text
        #get stop duration 
        if stop != stops[-1]:
            time_stop = stops_durations[stop_number].find_element_by_xpath('.//time').get_attribute('datetime')
            stop_number += 1
        else:
            time_stop = ''
        
        data_dict = {'CityOut':city_out, 'CityIn':city_in,'HourOut':hour_out,'HourIn':hour_in,'FlightDuration':duration,'Airplane':airplane,'StopDuration':time_stop}
        info_stops.append(data_dict)
    
    return info_stops
    
```

```python
def get_times(flight):
    
    #out hour
    hour_out = flight.find_element_by_xpath('.//div[@class = "departure"]/time').get_attribute('datetime')
    #in hour
    hour_in = flight.find_element_by_xpath('.//div[@class = "arrival"]/time').get_attribute('datetime')
    #duration
    duration = flight.find_element_by_xpath('.//span[@class = "duration"]/time').get_attribute('datetime')
    
    flight_info = {'OutHour':hour_out, 'InHour':hour_in,'Duration':duration}
    
    return flight_info
    
```

```python
def get_data(driver):
    flights = driver.find_elements_by_xpath('//li[@class = "flight"]')
    print('Exists {} number of flights'.format(len(flights)))
    print('Starting')
    info = []
    i=0
    for flight in flights:
        print(i)
        #total time
        gen_times = get_times(flight)
        #click Flight
        flight.find_element_by_xpath('./div[@class = "flight-container"]').click()
        #cost data
        cost_data = get_prices(flight)
        #click Flight
        flight.find_element_by_xpath('./div[@class = "flight-container"]').click()
        #click on the stops information
        flight.find_element_by_xpath('.//div[@class = "flight-summary-stops-description"]/button').click()
        stops_data = get_stops(flight)
        #click close 
        driver.find_element_by_xpath('//div[@class = "modal-header sc-dnqmqq cGfTsx"]//button[@class = "close"]').click()
        
        print(cost_data)
        info.append({'Costs':cost_data, 'Time':gen_times, 'Stops':stops_data})
        i+=1
    return info
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--incognito')
driver = webdriver.Chrome(ChromeDriverManager().install(), options = options)
driver.get(url)
delay = 10
try:
    flight_driver = WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,'//li[@class = "flight"]')))
    info_flights = get_data(driver)
except TimeoutException:
    print('the Page did not charge')
    info_flights = []
```
